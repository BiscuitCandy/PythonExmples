```python
import pandas as pd

df = pd.read_csv("weatherHistory.csv")
```

```python
df["Formatted Date"] = pd.to_datetime(df["Formatted Date"], utc = True)

df.head()
```

```python
df["Loud Cover"].value_counts()
```

```python
df.drop("Loud Cover", axis=1, inplace=True)
```

```python
df["Summary"].value_counts()
```

```python
from seaborn import heatmap
import matplotlib.pyplot as plt

heatmap(data = df.corr(), annot = True, fmt="0.01g")
plt.title("Correlations")
plt.savefig(fname = "Correlations.png", dpi = 500)
plt.show()
```

```python
df = df.set_index("Formatted Date")

req_attrs = ["Apparent Temperature (C)", "Humidity"]
re_data_mean = df[req_attrs].resample("MS").mean()
print(re_data_mean.shape)
re_data_mean.head()
```

```python
plt.scatter(re_data_mean["Apparent Temperature (C)"], re_data_mean["Humidity"])
plt.xlabel("Temp")
plt.ylabel("Humidity")
plt.title("ScatterPlot")
plt.show()
```

```python
from seaborn import lineplot

lineplot(data=re_data_mean)
```

```python
from seaborn import regplot

regplot(x=re_data_mean["Apparent Temperature (C)"], data=re_data_mean, y=re_data_mean["Humidity"])
plt.show()
```

```python
from seaborn import heatmap

heatmap(re_data_mean.corr(), annot = True, fmt="0.1g")
plt.title("Temp vs Hum")
plt.savefig("tmpvshum.png", dpi=250)
plt.show()
```

```python
from seaborn import FacetGrid, histplot

FacetGrid(data = df, hue="Summary", height=8).map(histplot,
                                 "Apparent Temperature (C)").add_legend()
plt.title("Weather wrt Tmp")
plt.show()
```

```python
from seaborn import FacetGrid, histplot

FacetGrid(df, hue="Summary", height=8).map(histplot, "Humidity").add_legend()
plt.title("Weather wrt Humidity")
plt.show()
```

```python
from seaborn import lineplot

months_colors = {1: ("Jan", "red"),
                 2 : ("Feb", "blue"),
                 3 : ("Mar", "green"),
                 4 : ("Apr", "black"),
                 5 : ("May", "brown"),
                 6 : ("Jun", "gray"),
                 7 : ("Jly", "yellow"),
                 8 : ("Aug", "purple"),
                 9 : ("Spt", "darkgreen"),
                 10 : ("Oct", "darkblue"),
                 11 : ("Nvm", "violet"),
                 12 : ("Dec", "orange")}

for i in range(1, 13) :
    lineplot(data = re_data_mean[re_data_mean.index.month==i]["Apparent Temperature (C)"], 
                color = months_colors[i][1], label = months_colors[i][0], marker = "o")

plt.title("TEMPERATURE")
plt.xlabel("Year")
plt.ylabel("Temp (C)")
plt.legend(fontsize="x-small")
```

```python
for i in range(1, 13) :
    lineplot(data = re_data_mean[re_data_mean.index.month==i]["Humidity"], 
                color = months_colors[i][1], label = months_colors[i][0], marker = "o")

plt.title("HUMIDITY")
plt.xlabel("Year")
plt.ylabel("Humidity")
plt.legend(fontsize="x-small")
```
